```python
#iris flower classification using ml
#libraries used

#aim-predict type of iris flower after training the model with iris dataset
#by using columns[parameters]-sepal length,petal length,sepal width and petal length
```

importing libraries and dataframe/dataset(df)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split

cols=['sepal-length','petal-length','sepal-width','petal-width','class']
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=cols)

df.head() #.head() shows the first 5 data frm dataset
```

```python
df.describe() #gives an analysis of the dataset like mean,etc...
```

```python
sb.pairplot(df,hue='class') #hue parameter is used to give color to the variables in plot for differentiaion
```

```python
data= df.values #data is a matrix and x and y are to slice dataset
X=data[:,0:4]
Y=data[:,4]
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print(X_test)
print(X_test.shape)
```

SUPPORT VECTOR MACHINE

```python
from sklearn.svm import SVC
model_svc=SVC()
model_svc.fit(X_train,Y_train)
```

```python
predictsvc= model_svc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictsvc)*100)
for i in range (len(predictsvc)):
  print(Y_test[i],predictsvc[i])
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model_lr=LogisticRegression()
model_lr.fit(X_train,Y_train)

```

```python
predictlr= model_lr.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictlr)*100)
for i in range(len( predictlr)):
  print(Y_test[i], predictlr[i])
```

decision tree

```python
from sklearn.tree import DecisionTreeClassifier
model_dtc=DecisionTreeClassifier()
model_dtc.fit(X_train,Y_train)

```

```python
predictdtc=model_dtc.predict(X_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictdtc))
```

```python
from sklearn.metrics  import classification_report
print(classification_report (Y_test,predictlr))
```
